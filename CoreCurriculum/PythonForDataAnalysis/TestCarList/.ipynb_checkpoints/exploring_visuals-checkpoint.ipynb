# Explorando com visualização
Usando os arquivos `clean_08.csv` e `clean_18.csv`

```python
# Imports e conjunto de dados
import pandas as pd

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

### Comparando as distribuções de classificação de emissão de gases do efeito estufa em 2008 e 2018.

```python
df_08['greenhouse_gas_score'].plot(kind='hist', title='Classificação de emissão de gases do efeito estufa 2008');
```

```python
df_18['greenhouse_gas_score'].plot(kind='hist', title='Classificação de emissão de gases do efeito estufa 2018');
```

### Como se alterou a distribuição de mpg de 2008 para 2018

```python
df_08['cmb_mpg'].plot(kind='hist', title='Distribuição de mpg 2008');
```

```python
df_18['cmb_mpg'].plot(kind='hist', title='Distribuição de mpg 2018');
```

### Correlação entre deslocamento cilindrada e mpg combinado

```python
df_08.plot(kind='scatter', x='cyl', y='cmb_mpg', title='Deslocamento cilindrada e mpg 2008');
```

```python
df_18.plot(kind='scatter', x='cyl', y='cmb_mpg', title='Deslocamento cilindrada e mpg 2018');
```

### Correlação entre classificação de emissão de gases do efeito estufa e mpg combinado

```python
df_08.plot(kind='scatter', x='greenhouse_gas_score', y='cmb_mpg', title='Classificação de efeito estufa e mpg 2008');
```

```python
df_18.plot(kind='scatter', x='greenhouse_gas_score', y='cmb_mpg', title='Classificação de efeito estufa e mpg 2018');
```
