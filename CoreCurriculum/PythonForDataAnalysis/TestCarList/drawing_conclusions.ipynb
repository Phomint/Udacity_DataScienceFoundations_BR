# Tirando conclusões
Usando os arquivos `clean_08.csv` e `clean_18.csv`.

```python
# Imports e conjunto de dados
import pandas as pd

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

### Existem mais modelos únicos usando fontes alternativas de combustível?

```python
df_08['fuel'].value_counts().plot(kind='bar', title='Tipos de combustíveis usados em 2008');
```

```python
df_18['fuel'].value_counts().plot(kind='bar', title='Tipos de combustíveis usados em 2018');
```

### Qual foi o nível de melhoria das classes de veículos no quisito "economia de combustível"?

```python
df_08.groupby('veh_class')['cmb_mpg'].mean().plot(kind='bar', ylim=(14, 30), title='Combinado de mpg dos veiculos em 2008');
```

```python
df_18.groupby('veh_class')['cmb_mpg'].mean().plot(kind='bar', ylim=(14, 30), title='Combinado de mpg dos veiculos em 2018');
```

### Quais são as características dos veículos SmartWay?

```python
smartway_08 = df_08.groupby('smartway').mean()
```

```python
smartway_08.plot(kind='bar', title='Tipos de carros com SmartWay 2008');
```

```python
smartway_18 = df_18.groupby('smartway').mean()
```

```python
smartway_18.plot(kind='bar', title='Tipos de carros com SmartWay 2018');
```
