# EDA com visualizações

```python
# Importando pacotes e carregando dataset
import pandas as pd

df_wine = pd.read_csv('winequality_edited.csv')
df_wine.head()
```

### Histogramas para diversas características

```python
df_wine['fixed acidity'].plot(kind='hist', title='Acidez');
```

```python
df_wine['total sulfur dioxide'].plot(kind='hist', title='Dióxido de Enxofre');
```

```python
df_wine['pH'].plot(kind='hist', title='pH');
```

```python
df_wine['alcohol'].plot(kind='hist', title='Álcool');
```

### Diagramas de dispersão da qualidade associada a diversas características

```python
df_wine.plot(kind='scatter', x='volatile acidity', y='quality', title='Acidez volátil');
```

```python
df_wine.plot(kind='scatter', x='residual sugar', y='quality', title='Açúcar residual');
```

```python
df_wine.plot(kind='scatter', x='pH', y='quality', title='pH');
```

```python
df_wine.plot(kind='scatter', x='alcohol', y='quality', title='Álcool');
```
