# Tirando conclusões usando groupby

```python
# Imports e carregando o dataset
import pandas as pd

df = pd.read_csv('winequality_edited.csv')
df.head()
```

```python
# Qualidade média de cada tipo de vinho (tinto e branco)

df.groupby('color', as_index=False)['quality'].mean()
```

```python
# Nível de acidez que recebe a maior avaliação média

df.describe()['pH']
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
```

```python
# Rótulos para os quatro níveis de acidez
bin_names = ['Alto', 'Moderado', 'Médio', 'Baixo']
```

```python
# Cria a coluna acidity_levels
df['acidity_levels'] = pd.cut(df['pH'], bin_edges, labels=bin_names)

# Verificação
df.head()
```

```python
# Qualidade média de cada nível de acidez com 'groupby'

df.groupby('acidity_levels', as_index=False)['quality'].mean()
```

```python
# Salvando alterações
df.to_csv('winequality_edited.csv', index=False)
```
